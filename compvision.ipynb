```python
import numpy as np
import matplotlib.pyplot as plt

image = np.full((16, 16, 3), 100, dtype='uint8')

colors = [
    [255, 0, 0],      # Red
    [255, 64, 0],     # Orange-Red
    [255, 128, 0],    # Orange
    [255, 192, 0],    # Gold
    [255, 255, 0],    # Yellow
    [128, 255, 0],    # Yellow-Green
    [0, 255, 0],      # Green
    [0, 255, 128],    # Spring Green
    [0, 255, 255],    # Cyan
    [0, 128, 255],    # Sky Blue
    [0, 0, 255],      # Blue
    [128, 0, 255],    # Violet
    [255, 0, 255],    # Magenta
    [255, 0, 128],    # Pink
    [255, 128, 128],  # Light Pink
    [255, 255, 255]   # White
]

for i in range(16):
    image[:, i] = colors[i]

plt.imshow(image)
plt.axis("off")
plt.show()
```
