INCORPORATING NEURAL NETWORK IN THE MODEL

```python
import pandas as pd
import numpy as np
import category_encoders as ce
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import joblib

# Load data
data = pd.read_csv('House_Rent_Dataset.csv')

# Split features and target
y = data['Rent']
x = data.drop(['Rent'], axis=1)
x = x.drop(['Posted On', 'Floor', 'Area Locality'], axis=1, errors='ignore')

# Train/test split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=42)

# Encode categorical columns (same as before)
encoder = ce.LeaveOneOutEncoder(random_state=42)
x_train_encoded = encoder.fit_transform(x_train, y_train)
x_test_encoded = encoder.transform(x_test)

# Scale features (same as before)
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train_encoded)
x_test_scaled = scaler.transform(x_test_encoded)

# Build the neural network
model = Sequential()
model.add(Dense(64, input_shape=(x_train_scaled.shape[1],), activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(1))  # single output, no activation = regression (predicting a number, not a class)

model.compile(optimizer='adam', loss='mse', metrics=['mae'])

model.summary()

# Train
history = model.fit(
    x_train_scaled, y_train,
    validation_data=(x_test_scaled, y_test),
    epochs=50,
    batch_size=16,
    verbose=1
)

# Evaluate
loss, mae = model.evaluate(x_test_scaled, y_test, verbose=0)
print("Test Mean Absolute Error (MAE):", mae)

# Save everything needed for future predictions
joblib.dump({
    'encoder': encoder,
    'scaler': scaler,
    'x_train_columns': x_train.columns
}, "house_rent_nn_preprocessor.pkl")

model.save("house_rent_nn_model.keras")

print("Neural network model and preprocessor saved.")
```

```python
# ---- Imports ----
import pandas as pd
import category_encoders as ce
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# ---- Load data ----
data = pd.read_csv('House_Rent_Dataset.csv')

y = data['Rent']
x = data.drop(['Rent'], axis=1)
x = x.drop(['Posted On', 'Floor', 'Area Locality'], axis=1, errors='ignore')

# ---- Train/test split ----
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=42)

# ---- Encode categorical (text) columns ----
encoder = ce.LeaveOneOutEncoder(random_state=42)
x_train_encoded = encoder.fit_transform(x_train, y_train)
x_test_encoded = encoder.transform(x_test)

# ---- Scale features ----
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train_encoded)
x_test_scaled = scaler.transform(x_test_encoded)

# ---- Build the neural network ----
model = Sequential()
model.add(Dense(64, input_shape=(x_train_scaled.shape[1],), activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(1))

model.compile(optimizer='adam', loss='mse', metrics=['mae'])

# ---- Train ----
model.fit(
    x_train_scaled, y_train,
    epochs=100,
    batch_size=32,
    validation_data=(x_test_scaled, y_test)
)

# ---- Predict on a few test samples ----
predictions = model.predict(x_test_scaled[1:5])
print(predictions)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Read dataset
data = pd.read_csv("wine.csv")

# Features and target
x = data.drop("quality", axis=1)
y = data["quality"]

# Convert labels from 3-8 to 0-5
y = y - 3

# Split dataset
x_train, x_test, y_train, y_test = train_test_split(
    x, y,
    test_size=0.25,
    random_state=42
)

# Create Neural Network
model = Sequential()

# Hidden Layer
model.add(Dense(64, activation="relu", input_shape=(x_train.shape[1],)))

# Output Layer (6 classes: 0,1,2,3,4,5)
model.add(Dense(6, activation="softmax"))

# Compile Model
model.compile(
    optimizer="adam", 
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"] #here optimizer decides whether the model is going to classify or it's going to perform regression. If it's classification, we use accuracy, if it's regression, we use mean squared error
)

# Train Model
model.fit(x_train, y_train, epochs=10, validation_data=(x_test, y_test))
history = model.fit(x_train, y_train, epochs=10, validation_data=(x_test, y_test))

# Test Accuracy
loss, accuracy = model.evaluate(x_test, y_test)

print("Accuracy:", accuracy)
```

TYPES= K-MEAN, HEIRARCHICAL CLUSTERING etc

step2= start creating a list
step3= generate a loop and check which number will go in which cluster(based on it's closeness to randomly assigned centres)

```python
li=[110,2,5,20,40,50,100,150]

#cluster centers (random selection)
cc1= 7
cc2= 45

clust1=[]
clust2=[]

for i in li:
    if abs(i-cc1)<abs(i-cc2):
        clust1.append(i)
    else:
        clust2.append(i)
print(clust1, clust2)
```
